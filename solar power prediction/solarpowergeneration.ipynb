```python
# Plot a histogram for the 'generated_power_kw' column
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'], kde=True, color='blue', bins=30)
plt.title('Distribution of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot missing values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Pairplot for selected features
selected_columns = ['temperature_2_m_above_gnd', 'relative_humidity_2_m_above_gnd', 'generated_power_kw']
sns.pairplot(df[selected_columns], diag_kind='kde', corner=True)
plt.show()

```

```python
# Visualize power distribution across day/night
g = sns.FacetGrid(df, col="day_or_night", height=5, aspect=1.2)
g.map(sns.histplot, "generated_power_kw", kde=True, color="blue")
plt.show()
```
